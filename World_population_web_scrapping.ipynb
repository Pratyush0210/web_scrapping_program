```python
import numpy as np
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.worldometers.info/world-population/population-by-country/"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, "lxml")
```

```python
table= soup.find("table", id = 'example2')
table_row = table.find_all("tr")
```

```python
len(table_row)
```

```python
col, contents = [], []

for i in range(len(table_row)):
    

    if i==0:
        col_list = table_row[i].find_all('th')
        for j in col_list:
            
            e = j.text
            if e == 'Yearly Change' or e == 'World Share':
                e = e + ' in %'
            col.append(e)
    else:
        row_list = table_row[i].find_all('td')
        content =[]
        x = 0
        for k in row_list:

            ele = k.text
            if ele == ' ' or ele == 'N.A.':
                pass
            elif x == 0  or x==9:
                ele = int(ele)

            elif x == 2 or x== 4 or x==6 or x == 7 or x==5:
                ele = int(''.join(ele.split(',')))

            elif x == 3 or x==11:
                ele = float(ele[:-2])
            
            elif x==8:
                ele = float(ele)
            
            elif x==10:
                ele = int(ele[:-2])
            x += 1
            content.append(ele)
        contents.append(content)

print(col)
```

```python
# print(contents)
```

```python
df = pd.DataFrame(contents, columns=col)
```
